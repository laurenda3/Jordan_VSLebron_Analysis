```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from Kaggle dataset folder
lebron_df = pd.read_csv('/kaggle/input/jordan-and-lebron-regular-season-stats/LJ_Reg_Sea - Sheet1.csv')
jordan_df = pd.read_csv('/kaggle/input/jordan-and-lebron-regular-season-stats/MJ_Reg_Sea - Sheet1.csv')

# Show first few rows
jordan_df.head()
```

```python
lebron_df.head()
```

```python
# Check for missing values and data types
print("Jordan's Dataset Info:")
print(jordan_df.info())
print("\nLeBron's Dataset Info:")
print(lebron_df.info())

# Look at summary statistics
print("\nJordan's Stats Summary:")
print(jordan_df.describe())

print("\nLeBron's Stats Summary:")
print(lebron_df.describe())
```

```python
# Basic Points Comparison
plt.figure(figsize=(8, 5))
plt.plot(jordan_df['PTS'], marker='o', label="Michael Jordan")
plt.plot(lebron_df['PTS'], marker='s', label="LeBron James")
plt.title("Points Per Game Over the Seasons")
plt.xlabel("Season")
plt.ylabel("Points Per Game")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Efficiency Comparison
labels = ['FG%', '3P%', 'FT%']
jordan_efficiency = [jordan_df['FG%'].mean(), jordan_df['3P%'].mean(), jordan_df['FT%'].mean()]
lebron_efficiency = [lebron_df['FG%'].mean(), lebron_df['3P%'].mean(), lebron_df['FT%'].mean()]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, jordan_efficiency, width, label='Michael Jordan')
plt.bar(x + width/2, lebron_efficiency, width, label='LeBron James')

plt.title('Efficiency Comparison (Career Averages)')
plt.ylabel('Percentage')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a summary of the selected stats
stats = ['REB', 'AST', 'STL', 'BLK']
jordan_avg = [jordan_df['REB'].mean(), jordan_df['AST'].mean(), jordan_df['STL'].mean(), jordan_df['BLK'].mean()]
lebron_avg = [lebron_df['REB'].mean(), lebron_df['AST'].mean(), lebron_df['STL'].mean(), lebron_df['BLK'].mean()]

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Set the width of the bars
bar_width = 0.35
index = range(len(stats))

# Plotting the bars for both players
bar1 = ax.bar(index, jordan_avg, bar_width, label="Jordan", color='b')
bar2 = ax.bar([i + bar_width for i in index], lebron_avg, bar_width, label="LeBron", color='r')

# Add labels and title
ax.set_xlabel('Statistic')
ax.set_ylabel('Average Per Game')
ax.set_title('Comparison of Rebounds, Assists, Steals, and Blocks')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(stats)

# Add the legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# True Shooting Percentage (TS%) Calculation:
# True Shooting Percentage accounts for field goals, 3-pointers, and free throws. The formula is:
# TS% = Points/ 2 × (Field Goals Attempted + 0.44 × Free Throws Attempted)
# This metric will give us a more accurate measure of their scoring efficiency.

# Calculate True Shooting Percentage (TS%) for Jordan
jordan_df['TS%'] = jordan_df['PTS'] / (2 * (jordan_df['FGA'] + 0.44 * jordan_df['FTA']))

# Calculate True Shooting Percentage (TS%) for LeBron
lebron_df['TS%'] = lebron_df['PTS'] / (2 * (lebron_df['FGA'] + 0.44 * lebron_df['FTA']))

# Display the TS% for both datasets
print(jordan_df[['PTS', 'FGA', 'FTA', 'TS%']].head())
print(lebron_df[['PTS', 'FGA', 'FTA', 'TS%']].head())
```

```python
print("Jordan TS% Mean:", jordan_df['TS%'].mean())
print("LeBron TS% Mean:", lebron_df['TS%'].mean())
```

```python
# Use games played (GS) as a proxy for season progression
jordan_df['Season'] = range(1, len(jordan_df) + 1)
lebron_df['Season'] = range(1, len(lebron_df) + 1)

# Plot TS% for both players
plt.figure(figsize=(10, 6))
plt.plot(jordan_df['Season'], jordan_df['TS%'], marker='o', label='Michael Jordan', color='red')
plt.plot(lebron_df['Season'], lebron_df['TS%'], marker='s', label='LeBron James', color='blue')

plt.title("True Shooting Percentage (TS%) Over Career")
plt.xlabel("Season (Based on Games Played)")
plt.ylabel("TS%")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Function to calculate Per 36 stats for relevant columns
def per_36(df):
    per_columns = ['PTS', 'AST', 'REB', 'STL', 'BLK', 'TOV']
    for col in per_columns:
        df[f'{col}_per_36'] = (df[col] / df['MIN']) * 36
    return df

# Apply the function to Jordan and LeBron DataFrames
jordan_df = per_36(jordan_df)
lebron_df = per_36(lebron_df)

# Display the updated DataFrames to check the new columns
print(jordan_df[['Season'] + [f'{col}_per_36' for col in ['PTS', 'AST', 'REB', 'STL', 'BLK', 'TOV']]])
print(lebron_df[['Season'] + [f'{col}_per_36' for col in ['PTS', 'AST', 'REB', 'STL', 'BLK', 'TOV']]])
```

```python

# List of stats to visualize
stats = ['PTS', 'AST', 'REB', 'STL', 'BLK', 'TOV']

# Create the plots
plt.figure(figsize=(14, 8))

for i, stat in enumerate(stats, 1):
    plt.subplot(2, 3, i)
    plt.plot(jordan_df['Season'], jordan_df[f'{stat}_per_36'], label=f"Michael Jordan {stat} per 36", color='red', marker='o')
    plt.plot(lebron_df['Season'], lebron_df[f'{stat}_per_36'], label=f"LeBron James {stat} per 36", color='blue', marker='o')
    plt.title(f'{stat} per 36')
    plt.xlabel('Season')
    plt.ylabel(f'{stat} per 36')
    plt.xticks(rotation=45)
    plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
